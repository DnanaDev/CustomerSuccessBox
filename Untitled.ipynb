### Multi-Variate Analysis

```python
## Getting the highest absolute correlations between features
# Get pair of features, find correlations for each pair, store in list along with feature names
# sort list to get features with the highest correlations
# initial implementation had two nested for loops, for 64*63*62, was too computationally expensive

from itertools import combinations

temp = train.drop(['Label', 'week_start_date'] , axis = 1).copy()
temp_corr = temp.corr()

corr_list = []
pair_feat = []
# Expect 64*64 outputs
for i, item in enumerate(list(combinations(temp.columns.to_list(),2))):
    pair_feat.append(item)
    
for item in pair_feat:
    corr_list.append([item[0], item[1], np.abs(temp_corr.at[item[0], item[1]])])

corr_array = np.array(corr_list)
corr_array = corr_array[corr_array[:,2].argsort()][::-1]
```

### Baseline Model with all features

```python
train.sort_index(inplace=True)
split = int(len(train) * .5)
x_train = train.iloc[: split+75]
x_test = train.iloc[split+75:]
```

```python
y_train = x_train['Label']
y_test = x_test['Label']
```

```python
print(y_train.value_counts()/len(y_train))
print(y_test.value_counts()/len(y_test))
```

```python
x_train.drop(['Label','week_start_date'], axis = 1, inplace=True)
x_test.drop(['Label', 'week_start_date'], axis = 1, inplace=True)
```

```python
class_weight = y_train.value_counts()[0]/y_train.value_counts()[1]
```

```python
clf = LogisticRegression( max_iter=100, class_weight='balanced')
clf_log = LogisticRegression(class_weight={0:1, 1:class_weight}, max_iter=100)
clf_rf = RandomForestClassifier(n_estimators=150, criterion='gini',
                                max_depth=15, class_weight={0:1, 1:class_weight})
xgb_clf = XGBClassifier(scale_pos_weight = class_weight)
```

```python
clf.fit(np.array(x_train), y_train)
clf_log.fit(np.array(x_train), y_train)
clf_rf.fit(np.array(x_train), y_train)
xgb_clf.fit(np.array(x_train), y_train)

print(f1_score(y_train, clf.predict(np.array(x_train))))
print(f1_score(y_train, clf_log.predict(np.array(x_train))))
print(f1_score(y_train, clf_rf.predict(np.array(x_train))))
print(f1_score(y_train, xgb_clf.predict(np.array(x_train))))
```

```python
y_pred = clf.predict(np.array(x_test))
y_pred_log = clf_log.predict(np.array(x_test))
y_pred_rf = clf_rf.predict(np.array(x_test))
y_pred_xgb = xgb_clf.predict(np.array(x_test))
```

```python
print(f1_score(y_test, y_pred))
print(f1_score(y_test, y_pred_log))
print(f1_score(y_test, y_pred_rf))
print(f1_score(y_test, y_pred_xgb))
```

```python
print(classification_report(y_test, clf.predict(np.array(x_test))))
print(classification_report(y_test, clf_log.predict(np.array(x_test))))
print(classification_report(y_test, clf_rf.predict(np.array(x_test))))
print(classification_report(y_test, xgb_clf.predict(np.array(x_test))))
```
